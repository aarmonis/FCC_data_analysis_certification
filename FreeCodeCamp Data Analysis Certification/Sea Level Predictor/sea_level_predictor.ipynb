# Import relevant libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

# Import dataframe with pd.read_csv()

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

# Create a scatter plot "CSIRO" vs "Year" and draw Lines of Best Fit (Since 1880, since 2000)

```python
#Preoare variables for lines of best fit
x = df.Year
y = df['CSIRO Adjusted Sea Level']
xx = df[df['Year']>=2000]['Year']
yy = df[df['Year']>=2000]['CSIRO Adjusted Sea Level']

#first the figure

plt.figure(1, figsize = (15,10))

#then the axes as a scatter subplot()
ax = plt.subplot()

ax.scatter(x,y)

#best fit 1
a1, b1, r_value, p_value, std_err = stats.linregress(x,y)

x1 = np.arange(x.min(), 2051)

y1 = a1 * x1 + b1

plt.plot(x1, y1, label = 'Best Fit Line 1')

#best fit 2
# y = a * x + b
line2 = stats.linregress(xx, yy)

a2 = line2.slope
b2 = line2.intercept

x2 = np.arange(xx.min(), 2051)
y2 = a2 * x2 + b2

plt.plot(x2, y2, label = 'Best Fit Line 2')

#cosmetics
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.grid()
```
